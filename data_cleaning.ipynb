```python
import pandas as pd
```

```python
def cleanData(fileName):
    df = pd.read_csv(fileName)
    df = df[df.Week.isnull() == False]
    df.rename(columns = {'Winner/tie': 'home_team', 'Loser/tie': 'away_team', 'PtsW': 'home_points',
                         'PtsL': 'away_points', 'YdsW': 'home_yards', 'TOW': 'home_to', 'YdsL': 'away_yards',
                         'TOL': 'away_to'}, inplace = True)
    condition = df['Away'] == '@'
    df.loc[condition, ['home_team', 'away_team', 'home_points', 'away_points', 'home_yards', 'home_to', 'away_yards',
                'away_to']] = df.loc[condition, ['away_team', 'home_team', 'away_points', 'home_points', 
                                                      'away_yards', 'away_to', 'home_yards', 'home_to']].values
    df = df.drop(columns = ["Day", "Date", "Time", "Away"])
    df.to_csv(fileName.split(".")[0] + "_clean.csv")
    return df
```

```python
cleanData('nfl2018.csv')
```

```python
cleanData('nfl2019.csv')
```

```python
cleanData('nfl2020.csv')
```
