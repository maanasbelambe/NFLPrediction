Libraries: sportsipy, sportsreference, pandas

Sport: NFL - Football

Input Statistics: offensive rating, defensive rating, strength of schedule, record in last four games, average points for, average points against

Input: Teams (with their stats), Spread, Over/Under Value

Output Statistics: Win/Loss, Difference in Score, Total Points Scored
- 3 Most Popular Bets: Moneyline (win/loss), Spread (difference in score), Over/Under (total points scored)

Output: Who Won, Result of Spread, Over or Under

```python
import pandas as pd
from sportsipy.nfl.teams import Teams
```

```python
def recordInLast4Games(team):
  wins = 0
  for game in team.schedule[team.games_played - 4 : team.games_played]:
    if game.result == "Win":
      wins += 1
    if game.result == "Tie": 
      wins += 0.5
  return wins/4
```

```python
from numpy.lib.function_base import average
teams = Teams()
list_teams = []
for team in teams:
  d = { "name": team.name, 
       "offensive_rating": team.offensive_simple_rating_system,
       "defensive_rating": team.defensive_simple_rating_system, 
       "strength_of_schedule": team.strength_of_schedule,
       "record_last_4": recordInLast4Games(team), 
       "avg_points_for": team.points_for / team.games_played,
       "avg_points_against": team.points_against / team.games_played, 
       "avg_point_differential": team.points_difference / team.games_played }
  list_teams.append(d)
df_teams = pd.DataFrame(list_teams)
```
