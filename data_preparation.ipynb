```python
import pandas as pd
import numpy as np
nfl2021 = pd.read_csv("nfl2021_clean.csv").drop(columns=["Unnamed: 0"])
```

```python
#home_win = 1 means home team won, home_win = 0 means home team lost
nfl2021["home_win"] = np.where(nfl2021["home_points"] > nfl2021["away_points"], 1, 0)
```

```python
nfl2021
```

```python
nfl2021_final = nfl2021[["Week", "home_team", "away_team", "home_win"]]
```

```python
def getWins(df, team, week):
    win_count = len(df[(((df.home_team == team) & (df.home_points > df.away_points)) | 
                   ((df.away_team == team) & (df.away_points > df.home_points))) &
                  (df.Week < week)])
    return win_count
```

```python
def getLosses(df, team, week):
    loss_count = len(df[(((df.home_team == team) & (df.home_points < df.away_points)) | 
                   ((df.away_team == team) & (df.away_points < df.home_points))) &
                  (df.Week < week)])
    return loss_count
```

```python
def getTies(df, team, week):
    tie_count = len(df[(((df.home_team == team) & (df.home_points == df.away_points)) | 
                   ((df.away_team == team) & (df.away_points == df.home_points))) &
                  (df.Week < week)])
    return tie_count
```

```python
def getPointsFor(df, team, week):
    df_home = df[(df.home_team == team) & (df.Week < week)]
    df_away = df[(df.away_team == team) & (df.Week < week)]
    total_points = df_home["home_points"].sum() + df_away["away_points"].sum()
    
    return total_points / (len(df_home) + len(df_away))
```

```python
def getPointsAgainst(df, team, week):
    df_home = df[(df.home_team == team) & (df.Week < week)]
    df_away = df[(df.away_team == team) & (df.Week < week)]
    total_points = df_home["away_points"].sum() + df_away["home_points"].sum()
    
    return total_points / (len(df_home) + len(df_away))
```

```python
def getYardsFor(df, team, week):
    df_home = df[(df.home_team == team) & (df.Week < week)]
    df_away = df[(df.away_team == team) & (df.Week < week)]
    total_yards = df_home["home_yards"].sum() + df_away["away_yards"].sum()
    
    return total_yards / (len(df_home) + len(df_away))
```

```python
def getYardsAgainst(df, team, week):
    df_home = df[(df.home_team == team) & (df.Week < week)]
    df_away = df[(df.away_team == team) & (df.Week < week)]
    total_yards = df_away["home_yards"].sum() + df_home["away_yards"].sum()
    
    return total_yards / (len(df_home) + len(df_away))
```

```python
def getTOfor(df, team, week):
    df_home = df[(df.home_team == team) & (df.Week < week)]
    df_away = df[(df.away_team == team) & (df.Week < week)]
    total_to = df_home["home_to"].sum() + df_away["away_to"].sum()
    
    return total_to / (len(df_home) + len(df_away))
```

```python
def getTOagainst(df, team, week):
    df_home = df[(df.home_team == team) & (df.Week < week)]
    df_away = df[(df.away_team == team) & (df.Week < week)]
    total_to = df_away["home_to"].sum() + df_home["away_to"].sum()
    
    return total_to / (len(df_home) + len(df_away))
```

```python
nfl2021_final[['home_wins', 'away_wins', 'home_losses', 'away_losses', 'home_ties', 'away_ties', 'home_points_for',
               'away_points_for', 'home_points_against', 'away_points_against', 'home_yards_for', 'away_yards_for',
               'home_yards_against', 'away_yards_against', 'home_to_for', 'away_to_for', 'home_to_against',
               'away_to_against']] = 0
```

```python
nfl2021_final
```

```python
for i in nfl2021_final.index:
    nfl2021_final['home_wins'][i] = getWins(nfl2021, nfl2021_final['home_team'][i], nfl2021_final['Week'][i])
    nfl2021_final['away_wins'][i] = getWins(nfl2021, nfl2021_final['away_team'][i], nfl2021_final['Week'][i])
    nfl2021_final['home_losses'][i] = getLosses(nfl2021, nfl2021_final['home_team'][i], nfl2021_final['Week'][i])
    nfl2021_final['away_losses'][i] = getLosses(nfl2021, nfl2021_final['away_team'][i], nfl2021_final['Week'][i])
    nfl2021_final['home_ties'][i] = getTies(nfl2021, nfl2021_final['home_team'][i], nfl2021_final['Week'][i])
    nfl2021_final['away_ties'][i] = getTies(nfl2021, nfl2021_final['away_team'][i], nfl2021_final['Week'][i])
    nfl2021_final['home_points_for'][i] = getPointsFor(nfl2021, nfl2021_final['home_team'][i], 
                                                       nfl2021_final['Week'][i])
    nfl2021_final['away_points_for'][i] = getPointsFor(nfl2021, nfl2021_final['away_team'][i],
                                                       nfl2021_final['Week'][i])
    nfl2021_final['home_points_against'][i] = getPointsAgainst(nfl2021, nfl2021_final['home_team'][i], 
                                                      nfl2021_final['Week'][i])
    nfl2021_final['away_points_against'][i] = getPointsAgainst(nfl2021, nfl2021_final['away_team'][i],
                                                      nfl2021_final['Week'][i])
    nfl2021_final['home_yards_for'][i] = getYardsFor(nfl2021, nfl2021_final['home_team'][i], 
                                                     nfl2021_final['Week'][i])
    nfl2021_final['away_yards_for'][i] = getYardsFor(nfl2021, nfl2021_final['away_team'][i], 
                                                 nfl2021_final['Week'][i])
    nfl2021_final['home_yards_against'][i] = getYardsAgainst(nfl2021, nfl2021_final['home_team'][i],
                                                     nfl2021_final['Week'][i])
    nfl2021_final['away_yards_against'][i] = getYardsAgainst(nfl2021, nfl2021_final['away_team'][i],
                                                     nfl2021_final['Week'][i])
    nfl2021_final['home_to_for'][i] = getTOfor(nfl2021, nfl2021_final['home_team'][i], nfl2021_final['Week'][i])
    nfl2021_final['away_to_for'][i] = getTOfor(nfl2021, nfl2021_final['away_team'][i], nfl2021_final['Week'][i])
    nfl2021_final['home_to_against'][i] = getTOagainst(nfl2021, nfl2021_final['home_team'][i],
                                                       nfl2021_final['Week'][i])
    nfl2021_final['away_to_against'][i] = getTOagainst(nfl2021, nfl2021_final['away_team'][i],
                                                       nfl2021_final['Week'][i])
```

```python
nfl2021_final['home_win_percentage'] = (nfl2021_final['home_wins'] + nfl2021_final['home_ties'] * 0.5
                                       ) / (nfl2021_final['home_wins'] + nfl2021_final['home_losses'] + 
                                            nfl2021_final['home_ties'])
```

```python
nfl2021_final['away_win_percentage'] = (nfl2021_final['away_wins'] + nfl2021_final['away_ties'] * 0.5
                                       ) / (nfl2021_final['away_wins'] + nfl2021_final['away_losses'] + 
                                            nfl2021_final['away_ties'])
```

```python
nfl2021_final = nfl2021_final.drop(['home_wins','away_wins','home_losses','away_losses','home_ties','away_ties'],
                                  axis = 1)
```

```python
nfl2021_final = nfl2021_final.reindex(columns=['Week','home_team','away_team', 'home_win_percentage', 
                                               'away_win_percentage', 'home_points_for',
                                               'away_points_for','home_points_against','away_points_against',
                                               'home_yards_for','away_yards_for','home_yards_against',
                                               'away_yards_against','home_to_for','away_to_for','home_to_against',
                                               'away_to_against','home_win'])
```
